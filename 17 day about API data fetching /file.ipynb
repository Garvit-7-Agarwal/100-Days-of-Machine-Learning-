##### URL = Protocol + Host + Path + Query Parameters

##### Example:
##### https://movie-database-api1.p.rapidapi.com/list_movies.json?limit=20&page=1

##### Protocol -> https
##### Host     -> movie-database-api1.p.rapidapi.com
##### Path     -> /list_movies.json
##### Query    -> limit=20&page=1

##### Headers are NOT part of the URL.
##### Example headers:
##### X-RapidAPI-Key : Your_API_Key
##### X-RapidAPI-Host: movie-database-api1.p.rapidapi.com

##### requests.get() uses:
##### requests.get(url, headers=headers, params=params)
