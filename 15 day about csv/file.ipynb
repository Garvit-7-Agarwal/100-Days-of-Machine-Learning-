 # 13.  Handling Dates

Converts date columns into DateTime format.

Allows date-related operations.

Benefits:

Extract month, year, day.

Perform date calculations easily.

# parse_dates Parameter in pandas

## Purpose

The `parse_dates` parameter is used in `pd.read_csv()` to convert date columns from **string/object type** to **datetime format** while reading the CSV file.

### Without parse_dates

```python
df = pd.read_csv("data.csv")
```

Date column datatype:

```python
object
```

### With parse_dates

```python
df = pd.read_csv("data.csv", parse_dates=["Date"])
```

Date column datatype:

```python
datetime64[ns]
```

This allows us to perform date-related operations easily.

---

# 1. List of Column Names

```python
pd.read_csv(
    "data.csv",
    parse_dates=["Date"]
)
```

### Meaning
- Convert the specified column into datetime format.

### Example

CSV:

```text
Date,Sales
2025-01-01,100
2025-01-02,120
```

Result:

```python
Date    -> datetime64[ns]
Sales   -> int64
```

---

# 2. Multiple Date Columns

```python
pd.read_csv(
    "data.csv",
    parse_dates=["Date","JoiningDate"]
)
```

### Meaning
- Convert multiple columns into datetime format.

### Example

CSV:

```text
Date,JoiningDate
2025-01-01,2024-06-01
2025-01-02,2024-07-01
```

Result:
- Both columns become datetime type.

---

# 3. List of Lists

Used when date information is split across multiple columns.

### Example

CSV:

```text
Year,Month,Sales
2025,01,100
2025,02,120
```

Code:

```python
pd.read_csv(
    "data.csv",
    parse_dates=[[0,1]]
)
```

### Meaning

- Combine column 0 (Year)
- Combine column 1 (Month)
- Create a single datetime column

Result:

```text
Year_Month  Sales
2025-01     100
2025-02     120
```

---

# 4. Dictionary

Used when combining columns and assigning a custom column name.

### Example

CSV:

```text
Year,Month,Sales
2025,01,100
2025,02,120
```

Code:

```python
pd.read_csv(
    "data.csv",
    parse_dates={"Date":[0,1]}
)
```

### Meaning

- Combine column 0 and column 1
- Convert them into a datetime column
- Name the new column "Date"

Result:

```text
Date      Sales
2025-01   100
2025-02   120
```

---

# Benefits of parse_dates

After converting to datetime:

```python
df["Date"].dt.year
df["Date"].dt.month
df["Date"].dt.day
```

Date calculations:

```python
df["Date"] - pd.Timestamp("2025-01-01")
```

Filtering:

```python
df[df["Date"] > "2025-01-01"]
```

---

# Most Common Usage

```python
df = pd.read_csv(
    "data.csv",
    parse_dates=["Date"]
)
```

This is the form used most frequently in real-world projects.

---

# Quick Summary

| Syntax | Purpose |
|----------|----------|
| `parse_dates=["Date"]` | Convert one date column |
| `parse_dates=["Date","JoiningDate"]` | Convert multiple date columns |
| `parse_dates=[[0,1]]` | Combine columns into one date column |
| `parse_dates={"Date":[0,1]}` | Combine columns and give custom name |

## Note

- By default, date columns are loaded as `object` (string).
- `parse_dates` converts them into `datetime64[ns]`.
- Datetime columns allow date filtering, date calculations, and extraction of year, month, and day.

# 14. Converters 
Used to apply custom transformations while loading data.

Example:

Convert full team names into short names.

Apply custom functions to specific columns.

````python
def rename(name):
    if name == "United States":
        return "cheaters"
    else:
        return name
````

````python
rename ("United States")
````

````python
pd.read_csv("netflix_titles.csv")
````

````python
pd.read_csv("netflix_titles.csv",converters={'country':rename})
````

# 16 Loading a huge data sets into chunks 

Used for very large datasets.

Loads data in smaller chunks instead of loading everything at once.

Helps when RAM is limited.

Benefits:

Handles huge datasets efficiently.

Operations can be performed chunk by chunk

````python
pd.read_csv("netflix_titles.csv")
````

````python
dfs = pd.read_csv("netflix_titles.csv",chunksize=2000)
````

````python
for chunks in dfs:
    print(chunks.shape)
````
