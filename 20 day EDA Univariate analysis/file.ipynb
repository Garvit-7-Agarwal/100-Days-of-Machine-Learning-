```python
import pandas as pd 
import seaborn as sns
df = pd.read_csv('train.csv')
```

```python
df.head()
```

# 1.) Categorical data 

## a.) countplot 

```python
sns.countplot(df, x="Survived")
```

```python
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

## b) Pie chart 

```python
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f') # autopct='%.2f' -> this helps used to calculate the percentage 
```

# 2). Numerical data 

## a) Histogram 

```python
import matplotlib.pyplot as plt 
plt.hist(df['Age'],bins=15)
```

## b) Distplot 

```python
sns.distplot(df['Age']) # Displot given the probability 
```

## c) BoxPlot

```python
sns.boxplot(df['Age'])
```
