# Working with dates 

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

## 1. Extract year 

```python
date['date_year'] = date['date'].dt.year

date.head()
```

## 2. Extract month 


```python
date['date_month_no'] = date['date'].dt.month
date.head()
```

```python
# if we want month name then 
date['date_month_no'] = date['date'].dt.month_name()
date.head()
```

## 3. Extract day no.

```python
date['date_day_no'] = date['date'].dt.day

date.head()
```

## # day of week ?

```python

date['date_dow'] = date['date'].dt.dayofweek
date.head()
```

## # day of week - name 

```python

date['date_dow_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

## # is weekend ?

```python

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Saturday','Sunday']),1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

## Extract week of the year 

```python
date['date_week'] = date['date'].dt.isocalendar().week

date.drop(columns=['product_id','city_id','orders']).head()
```

## Extract Quarter 

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

## Extract semester 

```python
date['semester'] = np.where(date['quarter'].isin([1,2]),1,2)

date.drop(columns=['product_id','city_id','orders']).head()
```

## Extract time elapsed between dats 

```python
import datetime 

today = datetime.datetime.today()
today
```

```python
today-date['date']
```

```python
# days passed 
(today-date['date']).dt.days
```

```python
# Month passed 
np.round((today - date['date']) / np.timedelta64(30, 'D'), 0)
```
